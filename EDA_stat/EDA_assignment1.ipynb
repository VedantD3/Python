# Identify significant variables using a correlation matrix

```python
matrix = df[numerical_df.columns.drop('Id')].corr().round(2)
```

```python
matrix.head()
```

```python
plt.figure(figsize=(25,25))
mask=np.triu(np.ones_like(matrix))
sns.heatmap(matrix,annot=True,annot_kws={"size": 9},mask=mask)
plt.show()
```

```python
matrix = matrix.unstack()
matrix.reset_index()
print(len(matrix))
matrix.head()
```

```python
# selecting variables with significant correlation
matrix = matrix[(abs(matrix)>= 0.7) & (matrix != 1)]
print(len(matrix))
print("---------------------------------------------------------------")
matrix.head()
```

```python
# getting unique pairs
correlated_var = matrix.index.unique()
```

```python
print(correlated_var)
print(len(correlated_var))
```

```python
significant_numVar = []
for ii in range(len(correlated_var)):
    print(ii)
    significant_numVar.append(correlated_var[ii][0])
    significant_numVar.append(correlated_var[ii][1])
    
```
